```python
from models.honeycomb import Honeycomb_Hamiltonian
from models.square import Square_Hamiltonian
from models.plotting import Plotting_Functions
```

The following is an example of using the imported square, honeycomb, and triangular lattice classes to generate some plots using the imported plotting functions.

```python
# Define Simulation Parameters

length = 10   # Lattice dimension
t = 1         # Hopping parameter
W = 0.0       # Disorder strength
phi = 0.5     # Magnetic flux per plaquette
q = 50        # Maximum denominator for phi values


# Initialize the Honeycomb

Honeycomb = Honeycomb_Hamiltonian(length, t, W, phi, q, save = True)

# Generate outputs of the honeycomb pass through to plot
honeycomb_outputs = Honeycomb.prepare_outputs()

# Initializing Honeycomb plotting
honeycomb_plot = Plotting_Functions(honeycomb_outputs, save = True)

# Plot the Eigenvalue spectrum of the Honeycomb
honeycomb_plot.plot_evals(save = True)

honeycomb_plot.plot_evec(save = True)

honeycomb_plot.plot_pr(save = True)

honeycomb_plot.plot_density_of_states(save = True)

Honeycomb.plot_hofstadter_butterfly(save = True)
```
