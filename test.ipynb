```python
from models.honeycomb import Honeycomb_Hamiltonian
from models.square import Square_Hamiltonian
from models.kagome import Kagome_Hamiltonian
from models.triangular import Triangular_Hamiltonian

from models.plotting import Plotting_Functions
```

The is notebook provides an example of how to import the lattice models containting butterfly plotting supports along with additional plotting functions. It's easy to set your paramters and call the relevant model and plotting function. All the plots here are set to save in the "plots" folder based on parameters. 

```python
square = Square_Hamiltonian(mode = 'butterfly', t=1, W=0.7, max_q=30, L = None, p=0, q=0, save=False)

square_outputs = square.prepare_outputs()
square.plot_hofstadter_butterfly(save = True)
```

```python
# Initialize the Honeycomb

Honeycomb = Honeycomb_Hamiltonian(mode = 'butterfly', t=1, W=0.6, max_q=30, L = None, p=0, q=0, save=False)

# Generate outputs of the honeycomb pass through to plot
honeycomb_outputs = Honeycomb.prepare_outputs()

# Initializing Honeycomb plotting
honeycomb_plot = Plotting_Functions(honeycomb_outputs, save = True)

Honeycomb.plot_hofstadter_butterfly(save = True)
```

```python
# Initialize the Honeycomb

Honeycomb = Honeycomb_Hamiltonian(mode = 'specific', t=1, W=0, max_q= None, L = 10, p=0, q=0, save=False)

# Generate outputs of the honeycomb pass through to plot
honeycomb_outputs = Honeycomb.prepare_outputs()

# Initializing Honeycomb plotting
honeycomb_plot = Plotting_Functions(honeycomb_outputs, save = True)

# Plot the Eigenvalue spectrum of the Honeycomb
honeycomb_plot.plot_evals(save = True)

honeycomb_plot.plot_evec(save = True)

honeycomb_plot.plot_pr(save = True)

honeycomb_plot.plot_density_of_states(save = True)
```

```python
# Initialize the Square
Square = Square_Hamiltonian(mode = 'specific', t=1, W=0, max_q= None, L = 10, p=0, q=0, save=False)

# Generate outputs of the square pass through to plot
square_outputs = Square.prepare_outputs()

# Initializing Honeycomb plotting
square_plot = Plotting_Functions(square_outputs, save = True)

# Plot the Eigenvalue spectrum of the Square
square_plot.plot_evals(save = True)

square_plot.plot_hall_conductance(save = True)
```

```python
# Initialize the Triangular

Tri = Triangular_Hamiltonian(mode = 'specific', t=1, W=0, max_q= None, L = 10, p=0, q=0, save=False)

# Generate outputs of the square pass through to plot
tri_outputs = Tri.prepare_outputs()

# Initializing Honeycomb plotting
tri_plot = Plotting_Functions(tri_outputs, save = True)

# Plot the Eigenvalue spectrum of the Square
tri_plot.plot_evals(save = True)
```
